```python
def plot_model_contributions(df, start_date=None, end_date=None, quantiles=None):
    """
    Plot comparison of GBLinear and LGBM contributions with quantile dropdown filter using Plotly.

    Args:
        df: DataFrame with model contribution columns
        start_date: Start date for filtering (str or datetime). If None, uses all data.
        end_date: End date for filtering (str or datetime). If None, uses all data.
        quantiles: List of quantile names to include in filter (e.g., ['P05', 'P50']). If None, uses all available.
    """
    # Filter dataframe by date range if specified
    df_plot = df.copy()
    if start_date is not None:
        start_dt = pd.to_datetime(start_date)
        if df_plot.index.tz is not None and start_dt.tz is None:
            start_dt = start_dt.tz_localize('UTC')
        df_plot = df_plot[df_plot.index >= start_dt]
    if end_date is not None:
        end_dt = pd.to_datetime(end_date)
        if df_plot.index.tz is not None and end_dt.tz is None:
            end_dt = end_dt.tz_localize('UTC')
        df_plot = df_plot[df_plot.index <= end_dt]

    # Get quantiles to plot
    if quantiles is None:
        quantiles = ['P05', 'P10', 'P30', 'P50', 'P70', 'P90', 'P95']

    # Create figure
    fig = go.Figure()

    # Add traces for each quantile
    for quantile in quantiles:
        gblinear_col = f'gblinear_quantile_{quantile}'
        lgbm_col = f'lgbm_quantile_{quantile}'

        # Add GBLinear trace
        fig.add_trace(
            go.Scatter(
            x=df_plot.index,
            y=df_plot[gblinear_col],
            name=f'GBLinear - {quantile}',
            mode='lines',
            line=dict(width=2, color='blue'),
            opacity=0.7,
            visible=(quantile == 'P50')  # Only P50 visible by default
            )
        )

        # Add LGBM trace
        fig.add_trace(
            go.Scatter(
            x=df_plot.index,
            y=df_plot[lgbm_col],
            name=f'LGBM - {quantile}',
            mode='lines',
            line=dict(width=2, color='#ff7f0e'),  # high-contrast orange for LGBM
            opacity=0.85,
            visible=(quantile == 'P50')  # Only P50 visible by default
            )
        )

        # Make the GBLinear trace (added just before) explicitly high-contrast too
        # fig.data[-2] refers to the GBLinear trace we added immediately prior
        if len(fig.data) >= 2:
            fig.data[-2].update(line=dict(color='#1f77b4', width=2), opacity=0.85)  # contrasting blue for GBLinear

    # Create dropdown buttons
    buttons = []
    for i, quantile in enumerate(quantiles):
        # Create visibility list - show only the selected quantile (both GBLinear and LGBM)
        visible = [False] * (len(quantiles) * 2)
        visible[i * 2] = True      # GBLinear for this quantile
        visible[i * 2 + 1] = True  # LGBM for this quantile

        buttons.append(
            dict(
                label=quantile,
                method='update',
                args=[
                    {'visible': visible},
                    {'title.text': f'Model Contributions - Quantile {quantile}'}
                ]
            )
        )

    # Update layout with dropdown menu
    # determine default active quantile (P50 if available)
    active_idx = quantiles.index('P50') if 'P50' in quantiles else 0

    fig.update_layout(
        legend=dict(
            orientation='v',
            x=1.02,
            xanchor='left',
            y=1.0,
            yanchor='top'
        ),
        updatemenus=[
            dict(
                type='buttons',
                buttons=buttons,
                direction='down',
                showactive=True,
                active=active_idx,  # Set the default active button
                x=1.02,
                xanchor='left',
                y=0.88,               # place beneath the legend
                yanchor='top',
                pad={'r': 10, 't': 10},
                bgcolor='lightgray',
                bordercolor='gray',
                borderwidth=1,
            )
        ],
        title=f"Model Contributions - Quantile {quantiles[active_idx]}",
        xaxis_title='Timestamp',
        yaxis_title='Contribution',
        hovermode='x unified',
        height=600,
        showlegend=True
    )

    fig.show()

# Example usage - plot all data
# plot_model_contributions(df_combined)
```
